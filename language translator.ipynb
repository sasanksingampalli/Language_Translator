```python
from googletrans import Translator
from ipywidgets import interact, widgets
from IPython.display import display
```

```python
def translate_text(text, target_language):
    translator = Translator()
    translated_text = translator.translate(text, dest=target_language)
    return translated_text.text
```

```python
def translate_ui():
    text_input = widgets.Text(description='Text to Translate:', value='what are you doing')
    language_dropdown = widgets.Dropdown(description='Target Language:', options=['English', 'French', 'German', 'Spanish','Hindi'], value='French')
    translated_output = widgets.Textarea(description='Translated Text:', disabled=True)

    def update_translation(change):
        translated_output.value = translate_text(text_input.value, get_language_code(language_dropdown.value))

    text_input.observe(update_translation, 'value')
    language_dropdown.observe(update_translation, 'value')

    display(text_input, language_dropdown, translated_output)
```

```python
def get_language_code(language):
    languages = {'English': 'en', 'French': 'fr', 'German': 'de', 'Spanish': 'es','Hindi':'hi'}
    return languages[language]
```

```python
def detect_language(text):
    translator = Translator()
    detected_language = translator.detect(text)
    return detected_language.lang
```

```python
def get_language_name(language_code):
    languages = {'en': 'English', 'fr': 'French', 'de': 'German', 'es': 'Spanish','hi':'Hindi'}
    return languages.get(language_code, 'Unknown')
```

```python
def detect_language_ui():
    text_input = widgets.Text(description='Text to Detect Language:', value='Hello, how are you?')
    detected_language_output = widgets.Text(description='Detected Language:', disabled=True)

    def update_detection(change):
        detected_language_code = detect_language(text_input.value)
        detected_language = get_language_name(detected_language_code)
        detected_language_output.value = detected_language

    text_input.observe(update_detection, 'value')

    display(text_input, detected_language_output)
```

```python
print("Translation Interface:")
translate_ui()
```

```python
print("\nLanguage Detection Interface:")
detect_language_ui()
```
